```python
def transform_bin(x):
    reg = 0
    exponent = 0
    while x > 0:
        if x % 2:
            reg += 10 ** exponent
        exponent += 1
        x = x // 2
    return reg

```

```python
possible_combinations = 1024 # possible numbers with 

combinations = [transform_bin(i) for i in range(possible_combinations)]
string_combinations = [f"{i:09d}" for i in combinations]

string_combinations
```
